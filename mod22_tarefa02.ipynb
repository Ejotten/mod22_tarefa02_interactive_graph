#### 2. Média móvel

1. Faça uma média móvel  em 90 dias dos valores de fechamento ajustados dos ativos que você escolheu
2. Utilizando a fução ```rolling()```, calcule o desvio pardão desses ativos ao longo do tempo, utilizando a mesma janela móvel do item acima
2. Monte um gráfico de linhas para as médias móveis dos ativos da sua base
4. Monte um gráfico de linhas para os desvios padrão em janela móvel
5. Compare os dois gráficos que você fez com o da tarefa anterior. Veja se eles apontam conclusões semelhantes quanto aos períodos de maior volatilidade (variação) dos ativos.

````python
# vamos selecionar apenas os valores ajustados de fechamento e calcular a amedia movel, os primeiros 90 linhas 
# não devem aparecer nada por conta da media movel. 
yf_adj= yf['Adj Close'].dropna()
yf_adj_90d = yf_adj.rolling(90).mean()
yf_adj_90d = yf_adj_90d.stack().reset_index()
yf_adj_90d.columns = ['Data','Ativo','valor']
yf_adj_90d
````

````python
yf_adj_dev = yf_adj.rolling(90).std()
yf_adj_dev = yf_adj_dev.stack().reset_index()
yf_adj_dev.columns = ['Data','Ativo','valor']
yf_adj_dev
````

````python
#grafico media movel 90 dias
fig = px.line(yf_adj_90d, x='Data', y="valor", color='Ativo', template='plotly_white')
fig.show()
````

````python
#grafico desvio padrão
fig = px.line(yf_adj_dev, x='Data', y="valor", color='Ativo', template='plotly_white')
fig.show()
````

## O desvio padrão é mais sensível e usado para entender quão volatil é uma ativo em um certo periodo de tempo, a media movel ele suaviza os valores do ativo para indicar tendencias e movimentos de medio longo prazo. 

#### 3. Gráfico de *candlestick*
1. Selecione o período dos últimos 60 dias (corridos) para um dos ativos da sua base
2. Monte um gráfico de *candlestick* para esses dados.

````python
#listas para criar o dicionario e puxar os dados
lista_indicadores = ['Adj Close', 'Close', 'High', 'Low', 'Open', 'Volume']
ativo = ['PETR4.SA']*len(lista_indicadores)
````

````python
colunas = list(zip(lista_indicadores, ativo))
````

````python
petr4 = yf.loc['09/20/2023':,colunas]
petr4.columns = lista_indicadores
````

````python
petr4
````

````python
graph = {
    'x': petr4.index,
    'open': petr4.Open,
    'close': petr4.Close,
    'high': petr4.High,
    'low': petr4.Low,
    'type': 'candlestick',
    'name': 'PETR4',
    'showlegend': True
}
````

````python
from plotly import graph_objects
from plotly.graph_objs import Layout

layout = Layout(
    paper_bgcolor='rgba(0,0,0,0)',
    plot_bgcolor='rgba(0,0,0,0)'
)

grafico = graph_objects.Figure(
    data = [graph],
    layout_title="Petrobras",
    layout=layout
)
grafico.update_xaxes(showgrid=True, gridwidth=1, gridcolor='LightGrey')
grafico.update_yaxes(showgrid=True, gridwidth=1, gridcolor='LightGrey')

grafico
````
